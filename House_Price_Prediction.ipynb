```python
def convertRange(x):
  temp = x.split("-")
  if len(temp) == 2:
    return (float(temp[0]) + float(temp[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
import numpy as np
def remove_outliers_sqft(dff):
  df_output = pd.DataFrame()
  for key,subdff in df.groupby('location'):
    m = np.mean(subdff.price_per_sqft)

    st = np.std(subdff.price_per_sqft)

    gen_df = subdff[(subdff.price_per_sqft > (m-st)) & (subdff.price_per_sqft <= (m+st))]
    df_output = pd.concat([df_output,gen_df],ignore_index= True)
  return df_output

df = remove_outliers_sqft(df)
df.describe()
```

```python
def bhk_outlier_remover(dff):
  exclude_indices = np.array([])
  for location,location_dff in dff.groupby('location'):
    bhk_stats={}
    for bhk,bhk_dff in location_dff.groupby('bhk'):
      bhk_stats[bhk] ={
          'mean' : np.mean(bhk_dff.price_per_sqft),
          'std' : np.std(bhk_dff.price_per_sqft),
          'count' : bhk_dff.shape[0]
      }

    for bhk,bk_df in location_dff.groupby('bhk'):
      stats = bhk_stats.get(bhk-1)
      if stats and stats['count']>5:
        exclude_indices = np.append(exclude_indices,bhk_dff[bhk_dff.price_per_sqft < (stats['mean'])].index.values)
  return dff.drop(exclude_indices,axis ='index')
```

```python
df = bhk_outlier_remover(df)
df
```

```python
x = df.drop(columns=['price'])
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=0)
```

```python
print(x_train.shape)
```

```python
print(x_test.shape)
```

Linear Regression

```python
column_trans = make_column_transformer((OneHotEncoder(sparse =False),['location']),remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr)
```

Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lasso = pipe.predict(x_test)
r2_score(y_test,y_pred_lasso)
```

Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans,scaler,ridge)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_ridge = pipe.predict(x_test)
r2_score(y_test,y_pred_ridge)
```

```python
print("No Regularization:",r2_score(y_test,y_pred_lr))
print("Lasso:",r2_score(y_test,y_pred_lasso))
print("Ridge:",r2_score(y_test,y_pred_ridge))
```
